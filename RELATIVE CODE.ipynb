```python
import cv2
import numpy as np
tlxx=0
tlyy=0
trxx=0
tryy=0
blxx=0
blyy=0
brxx=0
bryy=0

tmp = 0

# Reading the image
img = cv2.imread('KBs1.jpg')

# Converting the image into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Setting the threshold of the grayscale image
_, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Using the findContours() function
contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

i = 0

# Loop over the contours and detect shapes
for contour in contours:
    # Ignore the first contour as it represents the whole image
    if i == 0:
        i = 1
        continue
    
    # Approximate the contour to a polygon
    approx = cv2.approxPolyDP(contour, 0.01 * cv2.arcLength(contour, True), True)
    
    # Calculate the area of the shape
    area = cv2.contourArea(contour)
    
    # Skip shapes with an area less than or equal to 50
    if area < 4000:
        continue
    
    # Using the drawContours() function
    cv2.drawContours(img, [contour], 0, (0, 0, 255), 5)
    
    # Finding the center point of the shape
    M = cv2.moments(contour)
    if M['m00'] != 0.0:
        x = int(M['m10'] / M['m00'])
        y = int(M['m01'] / M['m00'])
        
        # Putting shape name at the center of each shape
        if len(approx) == 3:
            if tmp == 2:
                print('tri' + "(" + str(x) + "," + str(y) + ")")
                tlxx=x
                tlyy=y
            cv2.putText(img, str(tmp) + 'tri' + "(" + str(x) + "," + str(y) + ")", (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            tmp += 1
        elif len(approx) == 5:
            if tmp == 2:
                print('pent' + "(" + str(x) + "," + str(y) + ")")
                brxx=x
                bryy=y
            cv2.putText(img, str(tmp) + 'pent' + "(" + str(x) + "," + str(y) + ")", (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            tmp += 1
        elif len(approx) == 6:
            if tmp == 2:
                print('hex' + "(" + str(x) + "," + str(y) + ")")
                trxx=x
                tryy=y
            cv2.putText(img, str(tmp) + 'hex' + "(" + str(x) + "," + str(y) + ")", (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            tmp += 1
        elif len(approx) == 8:
            if tmp == 2:
                print('oct' + "(" + str(x) + "," + str(y) + ")")
                blxx=x
                blyy=y
            cv2.putText(img, str(tmp) + 'oct' + "(" + str(x) + "," + str(y) + ")", (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

# Save the image with detected shapes
cv2.imwrite("detected_shapes.jpg", img)
```
