```python
import cv2
import mediapipe as mp
import math

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.5)

# Initialize OpenCV video capture
cap = cv2.VideoCapture(1)  # Use the appropriate camera index (0, 1, etc.) if multiple cameras are available

while True:
    # Read frame from video capture
    ret, frame = cap.read()

    # Convert the frame to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame with MediaPipe Hands
    results = hands.process(frame_rgb)

    # Check if any hands are detected
    if results.multi_hand_landmarks:
        # Get the number of detected hands
        num_hands = len(results.multi_hand_landmarks)

        # Iterate through each detected hand
        for i, hand_landmarks in enumerate(results.multi_hand_landmarks):
            # Calculate the distance between INDEX_FINGER_PIP and THUMB_TIP
            index_pip_coords = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_PIP]
            thumb_tip_coords = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP]
            distance = math.sqrt((thumb_tip_coords.x - index_pip_coords.x)**2 + (thumb_tip_coords.y - index_pip_coords.y)**2)

            # Convert the index finger coordinates from normalized values to pixel coordinates
            height, width, _ = frame.shape
            cx, cy = int(index_pip_coords.x * width), int(index_pip_coords.y * height)

            # Check the distance threshold and print coordinates if applicable
            if (num_hands == 1) and distance < 0.1:
                # Print the coordinate of INDEX_FINGER_TIP
                index_tip_coords = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
                tip_cx, tip_cy = int(index_tip_coords.x * width), int(index_tip_coords.y * height)
                cv2.putText(frame, f"({tip_cx}, {tip_cy})", (tip_cx, tip_cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
            if num_hands == 2 and i == 0 and distance < 0.1:
                # Print the coordinate of INDEX_FINGER_TIP
                index_tip_coords = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
                tip_cx, tip_cy = int(index_tip_coords.x * width), int(index_tip_coords.y * height)
                cv2.putText(frame, f"({tip_cx}, {tip_cy})", (tip_cx, tip_cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
            if num_hands == 2 and i == 1 and distance < 0.1:
                # Print the coordinate of INDEX_FINGER_TIP
                index_tip_coords = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
                tip_cx, tip_cy = int(index_tip_coords.x * width), int(index_tip_coords.y * height)
                cv2.putText(frame, f"({tip_cx}, {tip_cy})", (tip_cx, tip_cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
            # Draw a circle at the tip of the index finger
            cv2.circle(frame, (cx, cy), 10, (0, 255, 0), -1)

    # Display the frame
    cv2.imshow('Frame', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close the OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
