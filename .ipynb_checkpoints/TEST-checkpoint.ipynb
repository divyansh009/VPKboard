```python
def draw_finger_angles(image, results, joint_list,lst,lst2,x1,y1,tot1,tot2):
    
    # Loop through hands
    for hand in results.multi_hand_landmarks:
        #Loop through joint sets 
        tmp=0
        for joint in joint_list:
            a = np.array([hand.landmark[joint[0]].x, hand.landmark[joint[0]].y]) # First coord
            b = np.array([hand.landmark[joint[1]].x, hand.landmark[joint[1]].y]) # Second coord(mid pt.)
            c = np.array([hand.landmark[joint[2]].x, hand.landmark[joint[2]].y]) # Third coord
            
            radians = np.arctan2(c[1] - b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
            angle = np.abs(radians*180.0/np.pi)
            
            if angle > 180.0:
                angle = 360-angle
            if hand == results.multi_hand_landmarks[0] and tmp==0:
                x1.append(angle)
                tot1.append(angle)
            elif len(results.multi_hand_landmarks)>1 and hand == results.multi_hand_landmarks[1] and tmp==0:
                y1.append(angle)
                tot2.append(angle)
            cv2.putText(image, str(round(angle, 2)), tuple(np.multiply(b, [640, 480]).astype(int)),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            if len(x1) > 30:
                x1.pop(0)
            if len(y1) > 30:
                y1.pop(0)
            tmp+=1
    return image
```

```python
#DTW FUNCTION:
def dtw(s, t, window):
    n, m = len(s), len(t)
    w = np.max([window, abs(n-m)])
    dtw_matrix = np.zeros((n+1, m+1))
    
    for i in range(n+1):
        for j in range(m+1):
            dtw_matrix[i, j] = np.inf
    dtw_matrix[0, 0] = 0
    
    for i in range(1, n+1):
        for j in range(np.max([1, i-w]), np.min([m, i+w])+1):
            dtw_matrix[i, j] = 0
    
    for i in range(1, n+1):
        for j in range(np.max([1, i-w]), np.min([m, i+w])+1):
            cost = abs(s[i-1] - t[j-1])
            # take last min from a square box
            last_min = np.min([dtw_matrix[i-1, j], dtw_matrix[i, j-1], dtw_matrix[i-1, j-1]])
            dtw_matrix[i, j] = cost + last_min
    return dtw_matrix

#FASTEST OVERALL
import numpy as np

def dtw(s, t, window):
    n, m = len(s), len(t)
    w = max(window, abs(n - m))
    w = max(w, abs(len(s) - len(t)))
    dtw_matrix = np.zeros((n + 1, m + 1))
    dtw_matrix[1:, 1:] = np.inf
    dtw_matrix[0, 0] = 0

    for i in range(1, n + 1):
        for j in range(max(1, i - w), min(m, i + w) + 1):
            cost = abs(s[i - 1] - t[j - 1])
            dtw_matrix[i, j] = cost + min(
                dtw_matrix[i - 1, j],  # insertion
                dtw_matrix[i, j - 1],  # deletion
                dtw_matrix[i - 1, j - 1]  # match/mismatch
            )

    return dtw_matrix[n, m]
```

```python
import cv2
import mediapipe as mp
import math
import numpy as np 

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=2,
                       min_detection_confidence=0.5,
                       min_tracking_confidence=0.5)

# Initialize arrays to store finger tip coordinates and angles
arr1 = []  # Left hand INDEX_FINGER_DIP
arr2 = []  # Right hand INDEX_FINGER_DIP
arr3 = []  # Left hand INDEX_FINGER_PIP
arr4 = []  # Right hand INDEX_FINGER_PIP
arr5 = []  # Left hand INDEX_FINGER_TIP
arr6 = []  # Right hand INDEX_FINGER_TIP
arr7 = []  # Left hand angle between DIP, PIP, and TIP
arr8 = []  # Right hand angle between DIP, PIP, and TIP

# Video capture settings
video_width = 640
video_height = 480

# Open video capture
cap = cv2.VideoCapture(1)
cap.set(3, video_width)
cap.set(4, video_height)

# Initialize MediaPipe drawing utilities
mp_drawing = mp.solutions.drawing_utils

while True:
    # Read frame from webcam
    ret, frame = cap.read()

    # Convert the image color space from BGR to RGB
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Flip the image horizontally for a mirrored display
    image = cv2.flip(image, 1)

    # Process the image with MediaPipe Hands
    results = hands.process(image)

    # Check if hands were detected
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            for idx, landmark in enumerate(hand_landmarks.landmark):
                # Get the coordinates of INDEX_FINGER_DIP (landmark index: 7)
                if idx == 7:
                    x = int(landmark.x * video_width)
                    y = int(landmark.y * video_height)

                    # Store the coordinates in the corresponding array
                    if hand_landmarks == results.multi_hand_landmarks[0]:
                        arr1.append((x, y))
                    elif hand_landmarks == results.multi_hand_landmarks[1]:
                        arr2.append((x, y))
                # Get the coordinates of INDEX_FINGER_PIP (landmark index: 6)
                elif idx == 6:
                    x = int(landmark.x * video_width)
                    y = int(landmark.y * video_height)

                    # Store the coordinates in the corresponding array
                    if hand_landmarks == results.multi_hand_landmarks[0]:
                        arr3.append((x, y))
                    elif hand_landmarks == results.multi_hand_landmarks[1]:
                        arr4.append((x, y))
                # Get the coordinates of INDEX_FINGER_TIP (landmark index: 8)
                elif idx == 8:
                    x = int(landmark.x * video_width)
                    y = int(landmark.y * video_height)

                    # Store the coordinates in the corresponding array
                    if hand_landmarks == results.multi_hand_landmarks[0]:
                        arr5.append((x, y))
                    elif hand_landmarks == results.multi_hand_landmarks[1]:
                        arr6.append((x, y))

            # Calculate the angle between DIP, PIP, and TIP
            if len(arr4) > 0 and len(arr5) > 0:
                dip_x, dip_y = arr4[-1]
                pip_x, pip_y = arr5[-1]
                tip_x, tip_y = arr6[-1]
                radians = np.arctan2(tip_y - pip_y, tip_x - pip_x) - np.arctan2(dip_y - pip_y, dip_x - pip_x)
                angle = np.abs(radians * 180.0 / np.pi)
                
                if angle > 180.0:
                    angle = 360 - angle
                ans = round(angle, 2)

                # Store the angle in the corresponding array
                if hand_landmarks == results.multi_hand_landmarks[0]:
                    arr7.append(ans)
                elif hand_landmarks == results.multi_hand_landmarks[1]:
                    arr8.append(ans)

            # Draw hand landmarks on the frame
            mp_drawing.draw_landmarks(
                image, hand_landmarks, mp_hands.HAND_CONNECTIONS)

            # Display angles on top of the screen
            if len(arr7) > 0:
                left_angle_text = "left = " + str(arr7[-1])
                cv2.putText(image, left_angle_text, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
            if len(arr8) > 0:
                right_angle_text = "right = " + str(arr8[-1])
                cv2.putText(image, right_angle_text, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Convert the image color space back to BGR for displaying
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Display the frame
    cv2.imshow('MediaPipe Hands', image)

    # Check for 'q' key press to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
